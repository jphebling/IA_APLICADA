```python
def escreve_texto_corretamente(texto):
    texto_novo = texto.upper()
    lista_de_palavras = texto_novo.split()
    nome_final = " ".join(lista_de_palavras)
    return nome_final
```

```python
lista_de_perguntas = [
    "De que é feito o Sol?",
    "De que é feito o planeta Saturno?",
    "Qual é a galáxia mais antiga já encontrada?",
    "Qual é a maior estrela já encontrada?",
    "Qual é a estrela mais próxima do Sol?"
]
```

```python
nova_lista_perguntas = []

with open("lista_perguntas_texto.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
      nova_lista_perguntas.append(linha.strip())
nova_lista_perguntas
```

```python
import os

from groq import Groq
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')

client_groq = Groq()
```

```python
def qa_perguntas(lista_de_perguntas):
  lista_de_dicionarios_de_respostas = []
  for numero, pergunta in enumerate(lista_de_perguntas):
    full_response_content = "" # Initialize an empty string to accumulate response
    completion = client_groq.chat.completions.create(
        model="openai/gpt-oss-20b", # Or any other Groq model you prefer, like 'llama3-8b-8192'
        messages=[
            {
                "role": "user",
                "content": f"Estou colocando em seguida uma pergunta. Preciso que analise o texto e responda em no máximo uma linha, sem ultrapassar este limite, garanta que seja apenas uma linha. Segue a pergunta: {pergunta}"
            }
        ],
        temperature=0.5,
        max_completion_tokens=8192,
        top_p=1,
        reasoning_effort="medium",
        stream=True,
        stop=None
    )

    # Iterate over the stream to get the full content
    for chunk in completion:
        if chunk.choices and chunk.choices[0].delta.content is not None:
            full_response_content += chunk.choices[0].delta.content

    # Replace non-breaking space with a regular space for better display
    processed_content = full_response_content.strip().replace('\u202f', ' ')
    lista_de_dicionarios_de_respostas.append({"pergunta": pergunta, "resposta": processed_content})
    df_perguntas_e_respostas = pd.DataFrame(lista_de_dicionarios_de_respostas)

  # Changed to return the list of responses for potential further use
  return df_perguntas_e_respostas
```

```python
df_qa = qa_perguntas(nova_lista_perguntas)
df_qa
```
